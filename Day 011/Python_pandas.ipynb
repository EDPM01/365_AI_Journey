## Pandas

Work with data

```python
#Import library
import pandas as pd
```

```python
# load data
dataset = pd.read_csv("wine.csv")
```

```python
#show dataset
dataset
```

```python
#show firsts
dataset.head(5)
```

```python
#show lasts
dataset.tail(3)
```

```python
#Dataset information
dataset.info()
```

```python
#dataset statistics
dataset.describe()
```

```python
#dataset dimensions
dataset.shape
```

```python
#Selecting Columns
dataset['V1']          # Single column
```

```python
dataset[['V1', 'V2']]  # Multiple columns
```

```python
#Filtering Rows
dataset[dataset['V1'] > 1]  # Rows where Column1 is greater than 1
```

```python
#Adding a New Column
dataset['new_column'] = dataset['V1'] + dataset['V2']
dataset
```

```python
#Dropping a Column
dataset.drop('new_column', axis=1, inplace=True)
dataset.head(5)
```

```python
#Sorting Data
dataset.sort_values(by='V2', ascending=True)
```

```python
# Grouping Data
dataset.groupby('V1').mean()
```

```python
# Unique
dataset['V1'].unique()
```

```python
#Handling Missing Values
dataset.isnull()               # Check for missing values

```

```python
dataset.isnull().sum()         # Count missing values per column
```

```python
dataset.fillna(0, inplace=True)  # Fill missing values with 0
```

```python
dataset.dropna(inplace=True)     # Drop rows with missing values
dataset.shape
```

```python
#histogram
dataset.V1.hist()
```

```python
dataset.V2.hist()
```

```python
dataset.V3.hist()
```

```python
#variable type
type(dataset)
```

```python
#matrix correlation
dataset.corr()
```

```python
dataset.corr(method='pearson')
```

```python
dataset.corr(method='spearman')
```

```python
#library visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matrix correlation
corr = dataset.corr()
```

```python
#show
corr
```

```python
#visualization
sns.heatmap(corr)
```

```python
#visualization, type of colors
sns.heatmap(corr, cmap='coolwarm')
```

```python
#visualization, type of colors, show values
sns.heatmap(corr, cmap='coolwarm', annot=True)
```

```python
#visualization, type of colors, show values, make figure of some size

plt.figure(figsize=(12,12))
sns.heatmap(corr, cmap='coolwarm', annot=True)
```
